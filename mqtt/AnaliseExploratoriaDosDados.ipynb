```python
def query(ID,p,dataset):
    df = pd.read_csv(dataset)
    query = df.iloc[len(df)-p:len(df)]
    ID = (query['ID'] == ID).tolist()
    return query[ID]
```

```python
query(1,500,"passagem.csv").plot()
```
